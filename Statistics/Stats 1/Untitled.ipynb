```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
from scipy.stats import norm
from scipy.stats import poisson
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15,3))


ns = [3,9,20]

for i in range(3):
    n, p = ns[i],0.3
    x = np.arange(0,n+1)
    rv = binom(n, p)
    ax[i].vlines(x, 0, rv.pmf(x), colors='k', linestyles='-', lw=1,
            label=f'binom({n},{p})')
    ax[i].plot(x, rv.pmf(x),'.', color='k')
    x = np.linspace(0,n,200)
    ax[i].plot(x, norm.pdf(x, loc=n*p, scale=(n*p*(1-p))**0.5), label=f'N({n*p:.2f}, {n*p*(1-p):.2f})', color='r')
    ax[i].legend()
plt.savefig('binom.png', bbox_inches='tight', dpi=200)
plt.show()
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15,3))

ns = [1,5,10]

for i in range(3):
    n = ns[i]
    x = np.arange(0,20)
    rv = poisson(n)
    ax[i].vlines(x, 0, rv.pmf(x), colors='k', linestyles='-', lw=1,
            label=f'poisson({n})')
    ax[i].plot(x, rv.pmf(x),'.', color='k')
    x = np.linspace(0,n,200)
    ax[i].legend()
plt.savefig('poisson.png', bbox_inches='tight', dpi=200)
plt.show()
```
